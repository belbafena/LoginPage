```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create the data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (reshaped to a 2D array)
y = np.array([1, 3, 2, 5, 4])                # Dependent variable

# Step 2: Create a linear regression model
model = LinearRegression()

# Step 3: Fit the model to the data
model.fit(X, y)

# Step 4: Get the coefficients
beta_0 = model.intercept_
beta_1 = model.coef_[0]

print(f"Intercept (beta_0): {beta_0}")
print(f"Slope (beta_1): {beta_1}")

# Step 5: Make predictions
y_pred = model.predict(X)

# Step 6: Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Original data')  # Plot the original data points
plt.plot(X, y_pred, color='red', label='Fitted line')  # Plot the regression line
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```
